# Investigate data

```python
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = sio.loadmat("../data/fashionMNIST.mat")

yTrain = data["yTrain"]
yTest = data["yTest"]
xTrain = data["xTrain"]
xTest = data["xTest"]
```

```python
labels_dict = dict()
for i, j in zip(range(1, 11), ['T-Shirt', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle Boot']):
    labels_dict[i] = j

labels_dict
```

Get a random selection of 16 images from the test data and display the images and their labels.
The images are only 28-by-28, so increase the size to 84-by-84 so they are easier to see.

```python
index = np.random.choice(yTest.shape[0], 16)

fig, axs = plt.subplots(3, 3)
k = 0
for i in range(3):
    for j in range(3):
        axs[i, j].imshow(xTest[:, :, index[k]], cmap="gray")
        axs[i, j].axis("off")
        axs[i, j].set_title(labels_dict[yTest[index[k], :][0]])
        k += 1

plt.show()
```

See the distribution of classes in the training and test data.

```python
print("Train data: ")
unique_values, counts = np.unique(yTrain, return_counts=True)
for value, count in zip(unique_values, counts):
    print(f"{labels_dict[value]} -> {count}")
```

```python
print("Test data: ")
unique_values, counts = np.unique(yTest, return_counts=True)
for value, count in zip(unique_values, counts):
    print(f"{labels_dict[value]} -> {count}")
```
