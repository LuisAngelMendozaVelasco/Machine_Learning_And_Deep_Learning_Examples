# View Augmented Images

```python
from keras.utils import image_dataset_from_directory
import matplotlib.pyplot as plt
from keras import Sequential
from keras.layers import RandomRotation
import numpy as np
import tensorflow as tf
```

```python
animals_dataset = image_dataset_from_directory("../data/Animals/", 
                                                seed=2,
                                                image_size=(224, 224))
```

An image dataset named animals_dataset has been created. It contains images of alpacas, cats and dogs.

```python
plt.figure()
for images, _ in animals_dataset.take(1):
    plt.imshow(images[0].numpy().astype("uint8"))
    plt.axis("off")
plt.show()
```

Possible augmentations include transformations like reflection, translation, and scaling.
Generally, we should choose an augmentation that is relevant to our data set. In this activity, we will rotate the images a random amount.

```python
augmentation = Sequential([RandomRotation(np.random.uniform(0, 1, 1)[0])])
```

When we create an augmented image datastore, we need to specify the augmenter.

```python
augmented_images = augmentation(images)
```

To display a transformed image, we need to extract one of the images from the dataset.

```python
plt.figure()
plt.imshow(augmented_images[0] / 255)
plt.axis("off")
plt.show()
```
