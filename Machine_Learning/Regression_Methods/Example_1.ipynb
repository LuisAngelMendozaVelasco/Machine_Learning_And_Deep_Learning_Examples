# Fitting a Line

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

sns.set_style("whitegrid")
```

```python
data = pd.read_excel("./data/data.xlsx", sheet_name='data')
data_train = pd.read_excel("./data/data.xlsx", sheet_name='dataTrain')
data_test = pd.read_excel("./data/data.xlsx", sheet_name='dataTest')
data.head()
```

```python
plt.figure()
plt.scatter(data.x, data.y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

The DataFrame data has been split into a training set and a test set. We can create a linear regression model using the [LinearRegression()](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html#sklearn-linear-model-linearregression) class.

```python
regressor = LinearRegression()
regressor.fit(data_train.x.to_frame(), data_train.y)
```

We can use the[ predict()](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html#sklearn.linear_model.LinearRegression.predict) function to predict the response on our test data.

```python
y_pred = regressor.predict(data_test.x.to_frame())
print("R^2:", regressor.score(data_test.x.to_frame(), data_test.y))

plt.figure()
plt.scatter(data.x, data.y)
plt.plot(data_test.x.to_frame(), y_pred, color='r')
plt.scatter(data_test.x, y_pred, color='r')
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
