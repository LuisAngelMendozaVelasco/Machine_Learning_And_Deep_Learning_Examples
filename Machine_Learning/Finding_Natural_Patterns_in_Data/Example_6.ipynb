# Options for k-Means Clustering

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

sns.set_style("whitegrid")
```

```python
X = pd.read_excel("./data/X.xlsx", sheet_name='X_2', header=None)
```

```python
kmeans = KMeans(n_clusters=3, n_init='auto')
kmeans.fit(X)
labels = kmeans.labels_
```

```python
plt.figure()
sns.scatterplot(x=X.iloc[:, 0], y=X.iloc[:, 1], hue=labels, palette="deep")
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, n_init=5, init="k-means++")
kmeans.fit(X)
labels = kmeans.labels_
```

```python
plt.figure()
sns.scatterplot(x=X.iloc[:, 0], y=X.iloc[:, 1], hue=labels, palette="deep")
plt.show()
```
