# Basketball Players

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
import numpy as np
from matplotlib.ticker import PercentFormatter
from sklearn.metrics import pairwise_distances
from  sklearn.manifold import MDS
from sklearn.preprocessing import StandardScaler

sns.set_style("whitegrid")
```

```python
data = pd.read_csv("./data/bball.csv")
data.pos = pd.Categorical(data.pos)
data.head()
```

```python
stats = data.iloc[:, [4, 5]+[i for i in range(10, data.shape[1])]]
stats.head()
```

```python
scaler = StandardScaler()
stats_normalized = scaler.fit_transform(stats)
```

The stats_normalized variable contains numeric statistics for several basketball players, normalized to have mean 0 and standard deviation 1.

```python
D = pairwise_distances(stats_normalized)
embedding = MDS(normalized_stress='auto', dissimilarity="precomputed", n_components=data.shape[1], random_state=0)
Y = embedding.fit_transform(D)
eigenvalues, _ = np.linalg.eig(Y.dot(Y.T))
```

```python
# Set figure and axis
fig, ax = plt.subplots()

# Plot bars (i.e. frequencies)
ax.bar(np.arange(1, 11), eigenvalues[:10].real)
ax.set_ylim(0, eigenvalues.real.sum()*1.05)
ax.set_title("Pareto Chart")
ax.set_xticks(np.arange(1, 11))

# Second y axis (i.e. cumulative percentage)
ax2 = ax.twinx()
ax2.plot(np.arange(1, 11), (eigenvalues[:10].real / eigenvalues.real.sum() * 100).cumsum(), color="red", marker="o")
ax2.axhline(90, color="orange", linestyle="dashed")
ax2.yaxis.set_major_formatter(PercentFormatter())
ax2.set_ylim(0, 100*1.05)

plt.show()
```

```python
fig = plt.figure()

ax = fig.add_subplot(projection='3d')
ax.scatter(Y[:, 0], Y[:, 1], Y[:, 2])

plt.show()
```

```python
pca = PCA(n_components=stats_normalized.shape[1])
Y = pca.fit_transform(stats_normalized)
variance_ratio = pca.explained_variance_ratio_
```

```python
# Set figure and axis
fig, ax = plt.subplots()

# Plot bars (i.e. frequencies)
ax.bar(np.arange(1, 11), variance_ratio[:10])
ax.set_ylim(0, variance_ratio.sum()*1.05)
ax.set_title("Pareto Chart")
ax.set_xticks(np.arange(1, 11))

# Second y axis (i.e. cumulative percentage)
ax2 = ax.twinx()
ax2.plot(np.arange(1, 11), (variance_ratio / variance_ratio.sum() * 100).cumsum()[:10], color="red", marker="o")
ax2.axhline(90, color="orange", linestyle="dashed")
ax2.yaxis.set_major_formatter(PercentFormatter())
ax2.set_ylim(0, 100*1.05)

plt.show()
```
