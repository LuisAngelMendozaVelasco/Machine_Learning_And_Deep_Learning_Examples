# Basketball Players

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

sns.set_style("whitegrid")
```

```python
data = pd.read_csv("./data/bball.csv")
data.pos = pd.Categorical(data.pos)
data.head()
```

```python
stats = data.iloc[:, [4, 5]+[i for i in range(10, data.shape[1])]]
stats.head()
```

```python
scaler = StandardScaler()
stats_normalized = scaler.fit_transform(stats)
```

```python
kmeans = KMeans(n_clusters=2, n_init='auto')
kmeans.fit(stats_normalized)
labels = kmeans.labels_
```

This code performs PCA and plots the transformed data by group.

```python
pca = PCA(n_components=stats_normalized.shape[1])
Y = pca.fit_transform(stats_normalized)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(Y[:, 0], Y[:, 1], Y[:, 2], c=labels, cmap='jet')
plt.show()
```
