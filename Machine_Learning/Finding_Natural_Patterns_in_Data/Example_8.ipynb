# Gaussian Mixture Models

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.mixture import GaussianMixture

sns.set_style("whitegrid")
```

```python
X = pd.read_excel("./data/X.xlsx", sheet_name='X_3', header=None)
```

We can use the [GaussianMixture()](https://scikit-learn.org/stable/modules/generated/sklearn.mixture.GaussianMixture.html#sklearn-mixture-gaussianmixture) class to fit a Gaussian mixture model to the data using k distributions.
We can specify the type of covariance matrix to use when fitting the distributions by setting the property "CovarianceType".

```python
model = GaussianMixture(n_components=3, covariance_type='diag')
labels = model.fit_predict(X)
```

We can visualize the groups by viewing the data in a scatter plot and assigning each group a different color.

```python
plt.figure()
sns.scatterplot(x=X.iloc[:, 0], y=X.iloc[:, 1], hue=labels, palette="deep")
plt.show()
```
