# Cross-Tabulation

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

sns.set_style("whitegrid")
```

```python
X = pd.read_csv("./data/data2.csv", header=None).loc[:, [0, 1]]
yCat = pd.read_csv("./data/data2.csv", header=None).loc[:, 2]
```

```python
kmeans = KMeans(n_clusters=2, n_init='auto')
kmeans.fit(X)
labels = kmeans.labels_
```

```python
plt.figure()
sns.scatterplot(x=X.loc[:, 0], y=X.loc[:, 1], hue=labels)
plt.show()
```

In some data sets, the observations already have a category associated with them. We can use the crosstab function to study the distribution of clusters across the original categories.

```python
counts = pd.crosstab(yCat, labels)
counts.columns.name = ''
counts.index.name = ''
counts
```

We can use the [heatmap()](https://seaborn.pydata.org/generated/seaborn.heatmap.html#seaborn-heatmap) function to visualize the counts.

```python
plt.figure()
sns.heatmap(counts, annot=True, fmt="")
plt.show()
```
