# Basketball Players

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.mixture import GaussianMixture
from sklearn.metrics import davies_bouldin_score

sns.set_style("whitegrid")
```

```python
data = pd.read_csv("./data/bball.csv")
data.pos = pd.Categorical(data.pos)
data.head()
```

```python
stats = data.iloc[:, [4, 5]+[i for i in range(10, data.shape[1])]]
stats.head()
```

```python
scaler = StandardScaler()
stats_normalized = scaler.fit_transform(stats)
```

```python
model = GaussianMixture(n_components=2, reg_covar=0.02)
labels = model.fit_predict(stats_normalized)
```

The DataFrame contains statistics for several basketball players. The players' positions are contained in data.pos.
The numeric statistics have been extracted and normalized, then used in a Gaussian mixture model (GMM) to divide the players into two groups.

```python
counts = pd.crosstab(data.pos, labels)
counts.columns.name = ''
counts.index.name = ''
counts
```

```python
plt.figure()
sns.heatmap(counts, annot=True, fmt="")
plt.show()
```

The Gaussian mixture model gmModel contains information about the underlying Gaussian distributions used to create the clusters. For example, we can extract the means of the distributions, which correspond to the centroids of the clusters.

```python
aux = pd.DataFrame(model.means_)
aux['labels'] = [0, 1]
aux.sort_values(by='labels', inplace=True)

plt.figure()
pd.plotting.parallel_coordinates(aux, "labels", colormap='jet')
plt.xticks(ticks=range(stats.shape[1]), labels=stats.columns, rotation=60)
plt.show()
```

```python
davies_bouldin_scores = {}
for n_cluster in range(2, 5):
    clusterer = GaussianMixture(n_components=2, reg_covar=0.02)
    cluster_labels = clusterer.fit_predict(stats_normalized)
    
    davies_bouldin_avg = davies_bouldin_score(stats_normalized, cluster_labels)
    davies_bouldin_scores[n_cluster] = davies_bouldin_avg
```

```python
print("Davies_bouldin evaluation:", davies_bouldin_scores)
print("Optimal number of clusters:", min(davies_bouldin_scores, key=davies_bouldin_scores.get))
```
