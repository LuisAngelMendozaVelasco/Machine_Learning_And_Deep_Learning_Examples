# k-Means Clustering

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

sns.set_style("whitegrid")
```

```python
X = pd.read_excel("./data/X.xlsx", sheet_name='X_1', header=None)
```

We can use the [KMeans()](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html#sklearn-cluster-kmeans) class to group data into k groups or clusters.

```python
kmeans = KMeans(n_clusters=3, n_init='auto')
kmeans.fit(X)
labels = kmeans.labels_
```

We can visualize the groups by viewing the data in a scatter plot and assigning each group a different color.

```python
plt.figure()
sns.scatterplot(x=X.iloc[:, 0], y=X.iloc[:, 1], hue=labels, palette="deep")
plt.show()
```
