# Explore Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
data = pd.read_csv("./data/bballData.csv")
data.head()
```

We can plot some results from the basketball player data to explore relationships between various statistics and player position.
For example, the [boxplot()](https://seaborn.pydata.org/generated/seaborn.boxplot.html#seaborn-boxplot) function can create a box plot of the height for each position.

```python
plt.figure()
sns.boxplot(y=data.height, x=data.pos)
plt.ylabel("Height (inches)")
plt.show()
```

It looks as though guards (G) are generally shorter than forwards (F) or centers (C). We can use [scatterplot()](https://seaborn.pydata.org/generated/seaborn.scatterplot.html#seaborn-scatterplot) to explore the relationship between two variables, grouped by position.

```python
plt.figure()
sns.scatterplot(x=data.rebounds, y=data.points, hue=data.pos)
plt.xlabel("Rebounds")
plt.ylabel("Points Scored")
plt.show()
```

We can see some groupings for different positions. However, the data's range makes it difficult to compare players, since most players are clustered tightly around the origin.
Some players played much more than others, leading to higher total points and rebounds. One way to account for this difference is to divide points and rebounds by the number of games played.
In the DataFrame data, the variable GP contains the number of games played. We can use element-wise division to calculate per game statistics.

```python
plt.figure()
sns.scatterplot(x=data.rebounds/data.GP, y=data.points/data.GP, hue=data.pos)
plt.xlabel("Rebounds per Game")
plt.ylabel("Points Scored per Game")
plt.show()
```

The data points are no longer clustered around the origin, but the data points for each position are still spread out. Another way to account for difference in play time is to divide by the number of minutes played, data.minutes.

```python
plt.figure()
sns.scatterplot(x=data.rebounds/data.minutes, y=data.points/data.minutes, hue=data.pos)
plt.xlabel("Rebounds per Game")
plt.ylabel("Points Scored per Game")
plt.show()
```
